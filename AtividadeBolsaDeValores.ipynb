```python
import pandas as pd
import matplotlib.pyplot as plt
base_df = pd.read_csv('/content/drive/MyDrive/base_cida/indexData.csv')
display(base_df)
```

```python
dataf1 = base_df[['Index', 'Close', 'Date']]
display(dataf1)
```

```python
dfremove = dataf1.loc[(base_df['Index']!='NYA')]
datanyadf = dataf1.drop(dfremove.index)
datanyadf = datanyadf[13940:]
display(datanyadf)
```

```python
figura = plt.figure(figsize=(15,5))
plt.plot(datanyadf['Date'],datanyadf['Close'],  label = 'NYA', color='b', ls = '--', lw = 2)
plt.legend(loc= 2, fontsize = 16)
plt.xlabel('Dia')
plt.ylabel('Cotação')
plt.title('Cotação da Bolsa de NY')
plt.show(figura)
```

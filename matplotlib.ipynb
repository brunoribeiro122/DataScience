```python
import matplotlib.pyplot as plt

x = [10, 20,30, 40]
y = [5, 10, 15, 10]
plt.plot(x, color = "red")
plt.plot(y, color = "blue")
plt.xlabel('grafico')
plt.show()
```

```python
plt.plot(x,y, label = 'dados')
plt.ylabel("eixo y")
plt.xlabel('eixo x')
plt.title('novo grafico')
plt.xticks([0,10,20,30,40,50,60])
plt.yticks([0,5,10,15,20])
plt.legend()
plt.show()
```

```python
plt.plot(x,y, linestyle='--', color='y', lw=10.0)
```

```python

plt.plot(x,y,ls='-', color ='#143252', lw=10.0)
plt.show()
```

```python
plt.scatter(x,y, color = "k")
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.bar(x,y, color = 'y')
plt.show()
```

```python
plt.boxplot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'ro')
plt.show()
```

```python
plt.scatter(x,y, color ='k', marker = '>')
plt.show()
```

```python
plt.plot(x,y)
plt.axis(xmin=10,xmax=60,ymin=0,ymax=16)
plt.show()
```

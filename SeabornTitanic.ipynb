```python
import seaborn as sns
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.info()
display(titanic) 
```

```python
titanic.isnull().sum()
```

```python
titanic.columns
```

```python
titanic.columns = [ 'sobreviventes', 'classe_passagem', 'sexo', 'idade', 'irmaos_a_bordo', 'pais/filhos_a_bordo', 'valor_passagem', 'local_embarque', 'classe', 'genero', 'homem', 'local_navio', 'torre_embarque', 'vivo', 'sozinho']
```

```python
titanic.columns
```

```python
sns.histplot(titanic['idade'], bins = 20, color = 'purple', kde = False)
```

```python
sns.histplot(titanic['idade'], bins = 20, color = 'red', kde = True).set(title="HistPlot")
```

```python
sns.displot(titanic['idade'], bins = 30, color = 'b', kde = True).set(title = 'Displot')
```

```python
sns.jointplot(data = titanic, x = 'idade', y = 'valor_passagem')
```

```python
sns.jointplot(data = titanic, x = 'pais/filhos_a_bordo', y = 'irmaos_a_bordo', hue = 'sobreviventes', kind = 'hist', bins = 7 )
```

```python
sns.jointplot(data = titanic, x = 'idade', y = 'valor_passagem', hue = 'sobreviventes')
```

```python
sns.jointplot(data = titanic, x = 'idade', y = 'classe_passagem', marker = '+', kind = 'reg', color = 'r')
```

```python
sns.lmplot(data = titanic, x = 'pais/filhos_a_bordo', y = 'idade', hue = 'sexo', markers = ['<', 'v'])
```

```python
titanics = titanic.drop(columns=['sozinho', 'homem'])
display(titanics)
```

```python
sns.pairplot(titanics)
```

```python
titanicos = titanic.astype(int)
sns.pairplot(titanicos)
```
